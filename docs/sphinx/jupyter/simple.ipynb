```python
import pandas as P
%matplotlib notebook
%pylab

from metadynamic import ResultReader, __version__

import matplotlib.pyplot as P
from mpl_toolkits.mplot3d import Axes3D                                         

print(f"metadynamic version: {__version__}")
```

```python
def plot3d(a,b,data, cmap="gist_gray_r", angle=-60, angle2=30, zticks=None, padx=30, pady=30, padz=30):
    top = data.max()
    amin, amax = a.min(), a.max()
    bmin, bmax = b.min(), b.max()
    fig = P.figure(figsize=(16,12))
    axe = fig.gca(projection="3d")
    axe.view_init(azim=angle, elev=angle2)
    axe.plot_surface(a, b, data, cmap=cmap,
                   linewidth=0, antialiased = True, alpha=0.9)
    axe.contour(a, b, data, offset = -top/2, zdir='z', cmap=cmap)
    axe.contour(a, b, data, offset = bmax, zdir='y', cmap=cmap)
    axe.contour(a, b, data, offset = amin, zdir='x', cmap=cmap)    
    axe.set_zlim(-top/2, top)
    axe.set_xlim(amin, amax)
    axe.set_ylim(bmin, bmax)
    axe.set_ylabel("length", labelpad=padx)
    axe.set_xlabel("time (s)", labelpad=pady)
    axe.set_zlabel("Population", labelpad=padz)
    if zticks:
        axe.set_zticks(zticks)
        
saveparams = {"bbox_inches":'tight', "pad_inches":0.0}
```

## Results: plotting

```python
figure()
plot(*res.x_y("a"), linewidth=2, color="tab:blue", label="$a$")
plot(*res.x_y("a", method="p2"), linewidth=1, color="tab:red", label="$a_2$")
plot(*res.x_y("a", method="p12"), linewidth=1, color="tab:green", label="$a_{12}$")


legend()
```

### Average and full results

```python
figure()
names = ["a", "b", "c", "ab", "bc", "abc"]
colors = ["blue", "orange", "green", "red", "purple", "pink"]

for name, color in zip(names, colors):
    plot(*res.x_y_proc(name), linewidth=0.2, color=f"tab:{color}")

for name, color in zip(names, colors):
    plot(*res.x_y(name), linewidth=2, color=f"tab:{color}", label=name)
    


legend()
xlim(0,end)
ylim(0)
```
